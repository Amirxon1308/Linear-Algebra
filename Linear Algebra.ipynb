## Kvadrat matritsa

Misol uchun quyidagi matritsa kvadrat matritsaga misol bo'ladi.

```python
# matritsa shaklida berilgan qiymatlarni listdan matritsaga o'tkazish

m_sqr = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
m_sqr_arr = np.array(m_sqr)
print(m_sqr_arr)
```

## Diagonal Matritsa

Diagonal matritsa deb diagonal shaklida joylashgan qiymatlari turlicha bo'lgan
va qolgan qiymatlari nolga teng bo'lgan matritsaga aytiladi.



```python
#Xuddi shu tarzdagi matritsani python orqali yaratish

m_diag = [[1, 0, 0], [0, 5, 0], [0, 0, 9]]
m_diag_np = np.matrix(m_diag)
print(m_diag_np)
```

```python
#diag() methodi yordamida matritsaning diagonalda joylashgan qiymatlarini ajratib olishimiz mumkin

m_sqr_mx = np.matrix('1 2 3; 4 5 6; 7 8 9')
diag = np.diag(m_sqr_mx)
print(diag)
```

```python
# diagonal shaklidagi matritsadan diagonal matritsani yaratish 
m_diag_np = np.diag(np.diag(m_sqr_mx))
print(m_diag_np)
```

# Teskari Matritsa

Teskari matritsa bu A matritsaga teskari bo'lgan A-1 bo'lgan matritsadir va ularning ko'paytmasi 1 ga teng:
    
    
E-bu birlik matritsadir. 

Hossa 1: Teskari matritsaning teskari matritsaning o'ziga teng:
    

```python
A = np.matrix('1. -3.; 2. 5.')
A_inv = np.linalg.inv(A)
A_inv_inv = np.linalg.inv(A_inv)

print(A)
```

```python
print(A_inv_inv)
```

Hossa 2: Aralashtirilgan matritsaning teskari va matritsa teskarisining aylantirilgani o'zgaro teng:
    

```python
#python:
A = np.matrix('1. -3.; 2. 5.')
L = np.linalg.inv(A.T)
R = (np.linalg.inv(A)).T
print(L)
```

```python
print(R)
```

Hossa 3: ikkita matritsaning ko'paytmasining teskarisi, ikkita matritsaning har birlarining teskarilari ko'paytmasi o'zaro teng
    

```python
A = np.matrix('1. -3.; 2. 5.')
B = np.matrix('7. 6.; 1. 8.')

L = np.linalg.inv(A.dot(B))
R = np.linalg.inv(B).dot(np.linalg.inv(A))

print(L)
```

```python
print(R)
```
